```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Qn3. Import data
path = r'C:\Users\Hiroshi\Documents\12-2023 Instacart Basket Analysis'
df_orders_products_combined = pd.read_pickle(os.path.join(path, '02 Data', '02 Prepared Data', 'orders_products_combined.pkl'))
df_prods = pd.read_csv(os.path.join(path, '02 Data', '02 Prepared Data', 'products_checked.csv'), index_col = False)
```

```python
# Qn5. Combine df_orders_products_combined and products data set (Test)
df_orders_products_combined.drop(columns = ['_merge','Unnamed: 0.1','Unnamed: 0'],inplace = True)
df_prods.drop(columns = ['Unnamed: 0'],inplace = True)
pd.merge(df_orders_products_combined,df_prods,on = ['product_id'], indicator = True, how = 'left')
```

```python
# Memory error troubleshooting: Change types for ords data set
df_orders_products_combined['order_id']=df_orders_products_combined['order_id'].astype('int32')
df_orders_products_combined['user_id'] = df_orders_products_combined['user_id'].astype('int32')
df_orders_products_combined['order_number']=df_orders_products_combined['order_number'].astype('int8')
df_orders_products_combined['order_dow']=df_orders_products_combined['order_dow'].astype('int8')
df_orders_products_combined['order_hour_of_day']=df_orders_products_combined['order_hour_of_day'].astype('int8')
df_orders_products_combined['days_since_prior_order']=df_orders_products_combined['days_since_prior_order'].astype('float16')
df_orders_products_combined['first_order']=df_orders_products_combined['first_order'].astype('int8')
df_orders_products_combined['product_id']=df_orders_products_combined['product_id'].astype('int32')
df_orders_products_combined['add_to_cart_order']=df_orders_products_combined['add_to_cart_order'].astype('int8')
df_orders_products_combined['reordered']=df_orders_products_combined['reordered'].astype('int8')
```

```python
# Memory error troubleshooting: Change types for prods data set
df_prods['product_id'] =df_prods['product_id'].astype('int32')
df_prods['aisle_id']=df_prods['aisle_id'].astype('int8')
df_prods['department_id']=df_prods['department_id'].astype('int8')
```

```python
# Qn5. Combine df_orders_products_combined and products data set
df_ords_prods_merge = df_orders_products_combined.merge(df_prods,on = ['product_id'], indicator = True, how = 'left')
```

```python
# Qn6. Confirm the results of the merge using the merge flag.
df_ords_prods_merge['_merge'].value_counts(dropna = False)
# This shows that it is not a full match. There are some products in the ords data that do not exist in the prods data.
```

```python
# Qn7. Export data.
df_ords_prods_merge.to_pickle(os.path.join(path, '02 Data','02 Prepared Data', 'ords_prods_merge.pkl'))
```
