```python
#Qn1. Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#Qn1. Import data
path = r'C:\Users\Hiroshi\Documents\12-2023 Instacart Basket Analysis'
ords_prods_cust_merge = pd.read_pickle(os.path.join(path, '02 Data', '02 Prepared Data', 'ords_prods_cust_merge_regcust.pkl'))
```

```python
# Qn2
# Plot bar chart
bar = ords_prods_cust_merge['order_dow'].value_counts().sort_index().plot.bar()
```

```python
#Turn outliers to NaNs
ords_prods_cust_merge.loc[ords_prods_cust_merge['prices'] >100, 'prices'] = np.nan
# Plot a histogram
hist = ords_prods_cust_merge['prices'].plot.hist(bins = 70)
plt.xlabel('Price ($)')
```

```python
# Plot a scatterplot
scatterplot = sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_cust_merge)
```

```python
# Create a list holding True/False values to the test np.random.rand()<=0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_cust_merge)) <= 0.7
dev
```

```python
# Keep only the columns required
df_sampling = ords_prods_cust_merge[['order_dow','prices']]
```

```python
# Store 30% of the sample in the dataframe small
small = df_sampling[~dev]
```

```python
# Plot a line chart
line = sns.lineplot(data = small, x = 'order_dow',y = 'prices')
plt.
```

```python
#Qn3. Plot a histogram of the "order_hour_of_day"
hist_hour_of_day = ords_prods_cust_merge['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram shows that it is the busiest from 9am-4pm. 12pm-7am is the time period with the least orders.

```python
# Qn4. Plot a bar chart from the "loyalty_flag"
bar_loyalty = ords_prods_cust_merge['loyalty_flag'].value_counts().sort_index().plot.bar()
plt.xticks(rotation = 0)
plt.xlabel('')
plt.ylabel('Number of orders')
plt.title('Order Distribution Among Customers of Varying Loyalty')
plt.tight_layout()
plt.show()
```

```python
# Export bar chart
bar_loyalty.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_loyalty.png'))
```

```python
# Qn6. Determine whether there’s a connection between age and family situation
df_sampling2 = ords_prods_cust_merge[['Age','n_dependants']]
```

```python
# Store 30% of the sample in the dataframe small
small2 = df_sampling2[~dev]
```

```python
# Plot the line chart.
line_dependants = sns.lineplot(data = small2, x = 'Age', y = 'n_dependants')
plt.title('Relationship Between Age and Number of Dependents')
plt.ylabel('Number of dependents')
plt.tight_layout()
plt.show()
```

```python
# Export line chart
line_dependants.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_dependants.png'))
```

The line chart shows that there is no connection between age and number of dependants an individual has.

```python
# Qn7. Explore whether there’s a connection between age and spending power (income).
scatterplot_income = sns.scatterplot(x = 'Age', y = 'income',data = ords_prods_cust_merge)
plt.ylabel('Income ($)')
plt.title('Relationship Between Age and Income')
plt.tight_layout()
plt.show()
```

We can see that there is no relationship between age and spending power.
