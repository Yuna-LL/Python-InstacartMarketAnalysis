```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Import data
path = r'C:\Users\Hiroshi\Documents\12-2023 Instacart Basket Analysis'
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', '02 Prepared Data', 'ords_prods_merge.pkl'))
```

```python
# Create subset
df = ords_prods_merge[:1000000]
```

```python
# Define function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply function to dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check value frequency
df['price_range'].value_counts(dropna=False)
```

```python
# Create price range lable using loc() for ords_prods_merge
ords_prods_merge.loc[ords_prods_merge['prices'] >= 15, 'price_range_loc'] = 'Higher-range product'
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] < 15, 'price_range_loc'] = 'Lower-range product'
```

```python
ords_prods_merge['price_range_loc'].value_counts(dropna=False)
```

```python
# Create busiest day lable using ords_prods_merge
result = []

for value in ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")

ords_prods_merge['busiest_day'] = result
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
# Qn2. Identify the two busiest days and two slowest days.
ords_prods_merge['order_dow'].value_counts(dropna=False)
# The busiest days are 0 and 1 (Saturday and Sunday)
# The slowest days are 3 and 4 (Tuesday and Wednesday).
```

```python
#Qn2. Create the two busiest days and two slowest days label.
result2 = []

for value in ords_prods_merge["order_dow"]:
  if (value == 0) or (value == 1):
    result2.append("Busiest days")
  elif (value == 3) or (value == 4):
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")

result2
```

```python
ords_prods_merge['busiest_2days'] = result2
```

```python
# Qn3. Check value frequency for busiest_2days column
ords_prods_merge['busiest_2days'].value_counts(dropna=False)
```

```python
# Qn4. Create new column to label the busiest period of day
# Check the total number of orders for each hour of day
ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
# For hour of day with >=2000,000 orders, label "Most orders"
# For hour of day with >=500,000 but <2000,000 orders, label "Average orders"
# For hour of day with <500,000 orders, label "Fewest orders"
```

```python
#Create the busiest period of day label.
result3 = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9,17]:
    result3.append("Most orders")
  elif value in [3,4,2,5,1,0,6,23]:
    result3.append("Fewest orders")
  else:
    result3.append("Average orders")

result3
```

```python
ords_prods_merge['busiest_period_of_day'] = result3
```

```python
# Qn5. Check value frequency for busiest_period_of_day column
ords_prods_merge['busiest_period_of_day'].value_counts(dropna=False)
```
